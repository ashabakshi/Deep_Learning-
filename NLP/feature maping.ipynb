```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import os
import cv2
```

```python
model = VGG16(weights='imagenet', include_top=False, pooling='avg')
```

```python
# Function to extract feature vector
def get_embedding(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return model.predict(x)[0]
```

```python
dataset_path = 'images/'  # folder with 10 images
image_files = os.listdir(dataset_path)
features = [get_embedding(os.path.join(dataset_path, f)) for f in image_files]
```

```python
# Query
query_img_path = 'Hrithik_Roshan.84.jpg'
query_feature = get_embedding(query_img_path)
```

```python
# Compute similarities
similarities = cosine_similarity([query_feature], features)[0]

# Sort by similarity
sorted_indices = np.argsort(similarities)[::-1]
for idx in sorted_indices:
    print(f"{image_files[idx]} - Similarity: {similarities[idx]:.2f}")
```
