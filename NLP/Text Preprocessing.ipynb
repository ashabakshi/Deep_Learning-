# Slang Translation

```python
f=open('slang.txt','r')
for i in f:
    print(i)
    if '=' in i:
        a=i.split('=')
        print(a)
#     break
```

```python
d={}
try:
    f=open('slang.txt','r')
    for i in f:
        if '=' in i:
            a=i.split('=')
            d[a[0]]=a[1].replace('\n','')
        elif ':' in i:
            a=i.split(':')
            d[a[0]]=a[1].replace('\n','')
        else:
            a=i.split('-')
            d[a[0]]=a[1].replace('\n','')
except:
    pass
```

```python
d
```

```python
len(d)+21
```

```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in d:
            new_text.append(d[w.upper()])
        else:
            new_text.append(w)
    return ' '.join(new_text)
            
```

```python
chat_conversion('asap')
```

# Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
# import nltk
# nltk.download()
```

```python
stopwords.words('english')
```

```python
def remove_stop_words(text):
    new_text=[]
    
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
            
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
            
```

```python
remove_stop_words('hello i am deepanshu')
```

```python
df['review'].head().apply(remove_stop_words)
```

# Handling Emoji's

```python
import re
```

```python
def remove_emoji(text):
    emoji_pattern=re.compile("["
                             u"\U0001F600-\U0001F64F" # emoticons
                             u"\U0001F300-\U0001F5FF" # symbols & pictographs
                             u"\U0001F680-\U0001F6FF" # transport and map symbols
                             u"\U0001F1E0-\U0001F1FF" # flags (ios)
                             u"\U00002702-\U000027B0"
                             u"\U000024C2-\U0001F251"
                             "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)
```

```python
remove_emoji("hello this is weired 🥲")
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
```

```python
WL=WordNetLemmatizer()
sentence='He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun.'
punctuation='?:!.,;'
sentence_words=nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuation:
        sentence_words.remove(word)
sentence_words
print("{0:20}{1:20}".format('Word','Lemma'))
for word in sentence_words:
    print("{0:20}{1:20}".format(word,WL.lemmatize(word,pos='v')))
    
```
