```python
import numpy as np
import pandas as pd                                                                                                                                                                                                                                                              
```

```python
df=pd.DataFrame({'Text. No.':['D1','D2','D3','D4'],'Text':['people watch movie','movie watch movie','people write comment','movie write comment']})
```

```python
df
```

```python
np.array(df['Text'])
```

# Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
bow=cv.fit_transform(df['Text'])
```

```python
# vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray()) # Document 1
print(bow[1].toarray()) # Document 2
print(bow[2].toarray()) # Document 3
print(bow[3].toarray()) # Documnet 4
```

```python
# for new sentence
cv.transform(['movie watch and write comment of movie ']).toarray() # prediction
```

```python
cv.vocabulary_
```

```python
# Binary Bag of Words
cv=CountVectorizer(    binary=True,
)
bow=cv.fit_transform(df['Text'])
```

```python
print(bow[0].toarray()) # Document 1
print(bow[1].toarray()) # Document 2
print(bow[2].toarray()) # Document 3
print(bow[3].toarray()) # Documnet 4
```

# N-grams

```python
# Bi-grams
cv=CountVectorizer(ngram_range=(2,2))
bow=cv.fit_transform(df['Text'])
```

```python
bow.toarray()
```

```python
# Tri-grams
cv=CountVectorizer(ngram_range=(3,3))
bow=cv.fit_transform(df['Text'])
print(cv.vocabulary_)
```

# Tf-Idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
tfidf.fit_transform(df['Text']).toarray()
```

```python
tfidf.vocabulary_
```

```python
tfidf.vocabulary_
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
X = cv.fit_transform(df['Text'])

print("Vocabulary:", cv.get_feature_names_out())
print("Term Frequency (Document-Term Matrix):\n", X.toarray())
```

```python
tfidf = TfidfVectorizer(use_idf=True)
X_tfidf = tfidf.fit_transform(df['Text'])

# TF-IDF values
print("TF-IDF Matrix:\n", X_tfidf.toarray())

# To see Term Frequency (before IDF is applied)
tfidf_no_idf = TfidfVectorizer(use_idf=False, norm=None)  # disables IDF
X_tf = tfidf_no_idf.fit_transform(df['Text'])
print("Term Frequency:\n", X_tf.toarray())

print("Vocabulary:", tfidf_no_idf.get_feature_names_out())
```
