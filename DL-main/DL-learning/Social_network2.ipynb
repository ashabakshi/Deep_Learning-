```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df['Gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
x=df.iloc[:,1:4]
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_scaled=st.fit_transform(x)
```

```python
x_scaled.shape
```

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(10,activation='relu',input_dim=3))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import time
```

```python
start=time.time()
history=model.fit(x_scaled,y,epochs=50,batch_size=400,validation_split=0.2)
end=time.time()
print(end-start)
```

```python
start=time.time()
history=model.fit(x_scaled,y,epochs=50,batch_size=1,validation_split=0.2)
end=time.time()
print(end-start)
```

```python
start=time.time()
history=model.fit(x_scaled,y,epochs=3,batch_size=120)
end=time.time()
print(end-start)
```
