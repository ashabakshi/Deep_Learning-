```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('diabetes (2).csv')
df
```

```python
X=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x_scaled=st.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32,validation_data=(x_test,y_test))
```

# How to Select a Appropriate Optimizer


```python
#pip install -U keras-tuner
```

```python
import kerastuner as kt
```

```python
def build_model(hp):

    model=Sequential()
    model.add(Dense(32,activation='relu',input_dim=8))
    model.add(Dense(1,activation='sigmoid'))


    optimizers=hp.Choice('optimizers',values=['sgd','adam','rmsprop','adadelta'])

    model.compile(optimizer=optimizers,loss='binary_crossentropy',metrics=['accuracy'])

    return model

    
    
```

```python
tuner=kt.RandomSearch(build_model,
                objective='val_accuracy',
                max_trials=5)
```

```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=100,initial_epoch=5,validation_data=(x_test,y_test))
```

# No of Nodes in a layer


```python
def build_model(hp):
    model=Sequential()
    units=hp.Int('units',min_value=8,max_value=128)
    model.add(Dense(units,activation='relu',input_dim=8))
    model.add(Dense(1,activation='sigmoid'))


    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])

    return model
```

```python
tuner=kt.RandomSearch(build_model,
                objective='val_accuracy',
                max_trials=5,
                directory='mydir',
                project_name='deepanshu')
```

```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=100,initial_epoch=5,validation_data=(x_test,y_test))
```

# How to select no of layers


```python
def build_model(hp):
    model=Sequential()
    model.add(Dense(85,activation='relu',input_dim=8))

    for i in range(hp.Int('num_layers',min_value=1,max_value=10)):
        model.add(Dense(85,activation='relu'))

    model.add(Dense(1,activation='sigmoid'))

    model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])

    return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=3,directory='mydir',project_name='verma')
```

```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=100,initial_epoch=5,validation_data=(x_test,y_test))
```

# All in all one model

```python
def build_model(hp):
    model=Sequential()
    counter=0

    for i in range(hp.Int('num_layers',min_value=1,max_value=10)):
        if counter==0:
            model.add(Dense(hp.Int('units'+str(i),min_value=8,max_value=128,step=8),
                            activation=hp.Choice('activation'+str(i),values=['relu','tanh','sigmoid']),
                            input_dim=8))

        else:
            model.add(Dense(hp.Int('units'+str(i),min_value=8,max_value=128,step=8),
                            activation=hp.Choice('activation'+str(i),values=['relu','tanh','sigmoid'])))


        counter+=1

    
    model.add(Dense(1,activation='sigmoid'))

    model.compile(optimizer=hp.Choice('optimizer',values=['sgd','adam','rmsprop','adadelta']),
            loss='binary_crossentropy',metrics=['accuracy'])
    #print('hello')

    return model
```

```python
tuner=kt.RandomSearch(build_model,
                objective='val_accuracy',
                max_trials=4,
                directory='mydir',
                project_name='final')
```

```python
tuner.search(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=200,initial_epoch=5,validation_data=(x_test,y_test))
```
