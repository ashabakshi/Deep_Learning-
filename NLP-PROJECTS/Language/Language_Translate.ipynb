```python
from deep_translator import GoogleTranslator

t = GoogleTranslator()

text = "my name is mohammad bilal"

GoogleTranslator(source="en",target='fr').translate(text)

def trans(text,target):
    trans_text  = GoogleTranslator(target=target).translate(text)
    return trans_text
    

while True:
    print("\n Language Translation Tool")
    input_text = input("Enter your text :")
    if input_text =='exit':
        break
    target_lang = input("To translate lang. (en,hi,ur,ta)")
    trans_text = trans(input_text,target_lang)
    print("Translation :",trans_text)
```
