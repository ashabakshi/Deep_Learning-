```python
doc= '\n'.join(df['title'].sample(1000))
# print(doc)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([doc.lower()])
```

```python
tokenizer.word_index
    
```

```python
print(len(tokenizer.word_index
         ))
```

```python
input_sequences=[] 
for i in df['title']:
    tokenized_sentence = tokenizer.texts_to_sequences([i])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])

```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
# padded_input_sequences[:,:-1]
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X[0].size
```

```python
y
```

```python
print(X.shape)
print(y.shape)
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=2890)
```

```python
print(y.shape)
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input

model = Sequential()
model.add(Input(shape=(26,)))
model.add(Embedding(2890, 100))
model.add(LSTM(150, return_sequences=True))
model.add(LSTM(150))
model.add(Dense(2890, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=1)
```

```python
for word,index in tokenizer.word_index.items():
    if index==pos:
        print(word)
```

```python
# multiple words


txt = 'nlp'
for i in range(5):
    # tokenize
    token_text = tokenizer.texts_to_sequences([txt])[0]
    # padding
    padded_token_text = pad_sequences([token_text],maxlen=27,padding='pre')
    # print(padded_token_text)
    # return the position that have a high probability
    pos = np.argmax(model.predict(padded_token_text))

    for word,index in tokenizer.word_index.items():
        if index==pos: 
            txt = txt + " " + word
            print(txt)
        
    
```
